```python
import torch
from torch import nn

fc1 = nn.Linear(1,3)
fc2 = nn.Linear(3,1)

x = torch.tensor([1.])
x = fc1(x)
print(x)
x = fc2(x)
print(x)

model = nn.Sequential(fc1, fc2)
x = torch.tensor([1.])
model(x)
```

```python
model = nn.Sequential(nn.Linear(2,5),
                      nn.Linear(5,10),
                      nn.Linear(10,10),
                      nn.Linear(10,3))

x = torch.randn(2)
model(x)

x = torch.randn(5,2) # 5개의 data, 근데 data 한 개가 두 개의 값을 가지고 있음

print(x)
print(model(x))
# 키, 몸무게 값을 다섯 사람 데이터에 대해서 통과 시킨것
```

```python
import torch
from torch import nn


class MyModel(nn.Module) :
    def __init__(self):
        super().__init__()

        self.fc1 = nn.Linear(2,5)
        self.fc2 = nn.Linear(5,10)
        self.fc3 = nn.Linear(10,3)

    def forward(self, x) :
        x = self.fc1(x)
        x = self.fc2(x)
        x = self.fc3(x)
        x = torch.heaviside(x, values = torch.tensor([0.5]))
        return x

model = MyModel()
x = torch.randn(5,2)
y = model(x)
print(y)
print(model)

print(model.fc1.weight)
print(model.fc1.bias)
```

```python
import torch
from torch import nn


class MyModel2(nn.Module) :
    def __init__(self):
        super().__init__()

        self.fcs = nn.Sequential(nn.Linear(2,5),
                                 nn.Linear(5,10),
                                 nn.Linear(10,3))

    def forward(self, x) :
        x = self.fcs(x)
        x = torch.heaviside(x, values = torch.tensor([0.5]))
        return x

model2 = MyModel2()
x = torch.randn(5,2)
y = model2(x)
print(y)

print(model2.fcs[0].weight)
print(model2.fcs[0].bias)
```

```python
# 파라미터 수 구하기

num = 0

for param in model.parameters():
    if param.requires_grad:
        num+= param.numel()

print(num)

num = 0

for param in model2.parameters():
    if param.requires_grad:
        num+= param.numel()

print(num)

```
